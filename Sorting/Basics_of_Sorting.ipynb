### Categories/ Classification of Sorting

#### 1. In-place and Not-in place Sorting
<b> In-place: </b>Algorithms do not require any extra space and sorting is said to happen in-place, or for example, within the array itself. <br>
These may require a small though non-constant extra space for its operation. Usually, this space is O(log n), though sometimes anything in O(n) (Smaller than linear) is allowed.<br>
Eg: Bubble, Selection, Insertion, Heap, Shell, Comb, Quick<br>
(QuickSort uses extra space for recursive function calls. It is called in-place according to broad definition as extra space required is not used to manipulate input, but only for recursive calls.)<br>

<b> Not-in place: </b>Algorithms which uses equal or more space than the elments being sorted.<br>
Eg: Merge(requires O(n) extra space).


#### 2. Stable and Not Stable Sorting
<b> Stable: </b>after sorting the contents, does not change the sequence of similar 
content in which they appear.<br>
Eg: Bubble, Insertion, Merge, Count,Radix





<b> Unstable: </b> after sorting the contents, changes the sequence of similar content 
in which they appear.<br>
Eg: Quick, Heap,Selection

Stability of an algorithm matters when we wish to maintain the sequence of original elements, like in a tuple. Eg: (student name,section) such various tuples so section 1 names will be first then that of section 2 and so on.<br>
A unstable sorting algorithm can be changed to stable sorting algorithm by changing the key comparison operation so that comparison of two keys considers position as a factor for objects with equal keys.<br>

#### 3.Adaptive and Non-Adaptive Sorting
<b> Adaptive: </b>  if it takes advantage of already 'sorted' elements in the list that is to be sorted. That is, while sorting if the source list has some element 
already sorted, adaptive algorithms will take this into account and will try not to re-order them.<br>
Eg: Bubble,Insertion, Quick<br>
<b> Non-Adaptive: </b>does not take into account the elements which are already sorted. They try to force every single element to be re-ordered to confirm their sortedness.<br>
Eg: Selection, Merge, Heap<br>

#### 4.Internal and External Sorting
<b> Internal: </b>use main memory exclusively during the sort,assumes high-speed random access to all memory,input data can be adjusted in the main memory at once,<br>
Example: Bubble, Insertion, Quick<br>
<b> External: </b>use external memory during the sorting,comparatively slower than internal sorting algorithms,input data cannot be adjusted in the memory entirely at once, it needs to be stored in a hard disk, floppy disk, or any other storage device.<br>
used where data is large, to select duplicate element, to to perform order by query<br>
Eg: merge, quick

#### 5.By Recursion
<b> Recursive: </b> Quick<br>
<b> Non- Recursive: </b> Selection, Insertion<br>
<b> Both: </b> Merge, Bubble

1. Increasing order:successive element is greater 
than the previous one.<br>
2. Decreasing order:successive element is less

3. Non-Increasing order:successive element is
